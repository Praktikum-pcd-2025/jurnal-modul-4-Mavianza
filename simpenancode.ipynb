```python
def sobel(img):
    gx = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
    gy = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
    
    x = conv(img, gx)
    x = np.sqrt(x*x)
    y = conv(img, gy)
    y = np.sqrt(y*y)
    
    gradient = abs(x)+abs(y)
    
    display(img, gradient, "Sobel")
    
    return gradient
```
